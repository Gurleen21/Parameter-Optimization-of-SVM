#### Splitting dataset into input and output features

```python
x=df.iloc[:,0:-1]
y=df['Class']
```

#### Scaling

```python
#Scaling input features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(x)
```

### Creating 10 Samples with 70-30 ratio of training and testing Set

```python
from sklearn.model_selection import train_test_split
samples = []
for i in range(10):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)
    samples.append((X_train, X_test, y_train, y_test))
```

```python
samples
```

### Creation of fitness function

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
result = pd.DataFrame(columns=['Sample', 'Best Accuracy', 'Best Kernel', 'Best Nu', 'Best Epsilon'])
```

```python
result
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
def fitnessFunction(kernel, C, gamma):
    itr=1000
    svm = SVC(kernel=kernel, C=C, gamma=gamma, max_iter=itr)
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    return accuracy_score(y_pred, y_test)
```

### Calling fitnes function for each sample

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
for i in range(len(samples)):
    best_score = 0
    best_C = 0
    best_gamma = 0
    for kernel in kernels:
        #print('Sample:', i, ' Kernel : ', kernel)
        X_train, X_test, y_train, y_test = samples[i]
        C = np.random.uniform(0, 10)
        # Randomly choose gamma
        gamma = np.random.uniform(0, 10)
        score = fitnessFunction(kernel, C, gamma)
        if score > best_score:
            best_score = round(score, 2)
            best_C = round(C, 2)
            best_gamma = round(gamma, 2)
            best_kernel = kernel
    print('Sample -',i,' Best Accuracy : ', best_score, ', Best kernel : ', best_kernel, ', Best Nu : ', best_C, ', Best Epsilon : ', best_gamma)
    result.loc[i] = [i, best_score, best_kernel, best_C, best_gamma]
```

### Result Table:

```python
print(result.to_markdown(showindex=False))
```

### Plotting Convergence Graph

```python
# Create convergence plot for best accuracy
from sklearn.model_selection import learning_curve

# Select sample with best accuracy
X_train, X_test, y_train, y_test = samples[result['Best Accuracy'].idxmax()]
```

```python
result['Best Accuracy'].idxmax()
```

```python
#Creating learning curve for best kernel
train_sizes, train_scores, test_scores = learning_curve(SVC(kernel=result['Best Kernel'].iloc[result['Best Accuracy'].idxmax()], 
                                                        C=result['Best Nu'].iloc[result['Best Accuracy'].idxmax()],
                                                        gamma=result['Best Epsilon'].iloc[result['Best Accuracy'].idxmax()],
                                                        max_iter = 1000), X_train, y_train, cv=10, scoring='accuracy', n_jobs=-1,
                                                        train_sizes = np.linspace(0.01, 1.0, 50))
```

```python
train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, label='Training Score')
plt.plot(train_sizes, test_mean, label='Cross-Validation Score')
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.title('Convergence Graph')
plt.legend(loc="best")
plt.show()
```
